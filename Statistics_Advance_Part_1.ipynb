3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(p, x):
    """
    Calculate the Probability Distribution Function (PDF) of a Bernoulli distribution.

    Parameters:
        p (float): Probability of success (0 <= p <= 1).
        x (int): Random variable (can be 0 or 1).

    Returns:
        float: The probability associated with the value of x.
    """
    if x not in [0, 1]:
        raise ValueError("The random variable x must be either 0 or 1.")
    if not (0 <= p <= 1):
        raise ValueError("The probability p must be between 0 and 1.")

    # Bernoulli PDF formula
    return p if x == 1 else 1 - p

# Example usage
p = 0.7  # Probability of success
x = 1    # Random variable
probability = bernoulli_pdf(p, x)
print(f"The probability of x = {x} in the Bernoulli distribution with p = {p} is: {probability}")
```

9. Write a Python function to calculate Z-scores from a dataset and plot them.


```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(dataset):
    """
    Calculate the Z-scores for a given dataset.

    Parameters:
        dataset (array-like): The dataset for which Z-scores are calculated.

    Returns:
        array-like: Z-scores for the dataset.
    """
    mean = np.mean(dataset)  # Calculate the mean of the dataset
    std_dev = np.std(dataset)  # Calculate the standard deviation of the dataset
    z_scores = (dataset - mean) / std_dev  # Compute Z-scores
    return z_scores

def plot_z_scores(z_scores):
    """
    Plot the Z-scores using a histogram.

    Parameters:
        z_scores (array-like): Z-scores to be plotted.
    """
    plt.hist(z_scores, bins=30, color='skyblue', edgecolor='blue', alpha=0.7, label='Z-scores')
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.title('Histogram of Z-scores')
    plt.legend()
    plt.grid(alpha=0.7, linestyle='--')
    plt.show()

# Example dataset
dataset = np.random.normal(loc=100, scale=15, size=1000)  # Generate a random normal dataset

# Calculate Z-scores and plot them
z_scores = calculate_z_scores(dataset)
plot_z_scores(z_scores)
```

10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a non-normal distribution (e.g., exponential distribution)
original_data = np.random.exponential(scale=2, size=1000)

# Function to compute sample means
def compute_sample_means(data, sample_size, num_samples):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(data, size=sample_size, replace=True)
        sample_means.append(np.mean(sample))
    return sample_means

# Parameters
sample_size = 30  # Size of each sample
num_samples = 1000  # Number of samples to draw

# Compute the sample means
sample_means = compute_sample_means(original_data, sample_size, num_samples)

# Plot the original non-normal distribution
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(original_data, bins=30, color='skyblue', edgecolor='blue', alpha=0.7, density=True)
plt.title('Original Non-Normal Distribution (Exponential)')
plt.xlabel('Value')
plt.ylabel('Density')

# Plot the distribution of the sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='green', edgecolor='darkgreen', alpha=0.7, density=True)
plt.title('Distribution of Sample Means (CLT)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, dataset):
    """
    Calculate the Z-score for a given data point.

    Parameters:
        data_point (float): The data point to calculate the Z-score for.
        dataset (array-like): The dataset for comparison.

    Returns:
        float: The calculated Z-score.
    """
    mean = np.mean(dataset)  # Calculate the mean of the dataset
    std_dev = np.std(dataset)  # Calculate the standard deviation of the dataset
    z_score = (data_point - mean) / std_dev
    return z_score

def compare_to_standard_normal(z_score):
    """
    Compare the Z-score to the standard normal distribution.

    Parameters:
        z_score (float): The Z-score to compare.

    Displays:
        A plot showing where the Z-score falls within the standard normal distribution.
    """
    # Generate x values and PDF for the standard normal distribution
    x = np.linspace(-4, 4, 1000)
    pdf = norm.pdf(x)

    # Plot the standard normal distribution
    plt.plot(x, pdf, label='Standard Normal Distribution', color='blue')
    plt.fill_between(x, pdf, alpha=0.3, color='skyblue')

    # Highlight the Z-score on the plot
    plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')
    plt.title('Comparison with Standard Normal Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(alpha=0.7, linestyle='--')
    plt.show()

# Example usage
dataset = [10, 12, 14, 16, 18, 20, 22]  # Example dataset
data_point = 16  # Example data point
z_score = calculate_z_score(data_point, dataset)
print(f"The Z-score for the data point {data_point} is: {z_score:.2f}")
compare_to_standard_normal(z_score)
```

15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std):
    """
    Perform Z-test for a sample dataset.

    Parameters:
        sample (array-like): The sample dataset.
        population_mean (float): The mean of the population.
        population_std (float): The standard deviation of the population.

    Returns:
        float: Z-statistic value.
        float: P-value (two-tailed test).
    """
    # Calculate sample mean
    sample_mean = np.mean(sample)

    # Calculate Z-statistic
    n = len(sample)
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Calculate p-value for a two-tailed test
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

    return z_statistic, p_value

# Example usage
sample_data = [18, 20, 22, 24, 26]  # Sample dataset
population_mean = 25  # Hypothesized population mean
population_std = 5  # Population standard deviation

z_stat, p_val = z_test(sample_data, population_mean, population_std)
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_val:.4f}")

# Decision based on p-value
alpha = 0.05  # Significance level
if p_val < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

16. Create a confidence interval for a dataset using Python and interpret the result.


```python
import numpy as np
from scipy.stats import norm

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for a given dataset.

    Parameters:
        data (array-like): The dataset for which the confidence interval is calculated.
        confidence_level (float): The desired confidence level (default is 0.95).

    Returns:
        tuple: Lower and upper bounds of the confidence interval.
    """
    # Calculate the mean and standard error of the sample
    sample_mean = np.mean(data)
    std_error = np.std(data, ddof=1) / np.sqrt(len(data))

    # Find the critical Z-value for the confidence level
    z_value = norm.ppf((1 + confidence_level) / 2)

    # Calculate the margin of error
    margin_of_error = z_value * std_error

    # Determine the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example dataset
data = [18, 20, 22, 24, 26, 28, 30, 32, 34, 36]

# Calculate the confidence interval
confidence_level = 0.95  # 95% confidence level
lower, upper = calculate_confidence_interval(data, confidence_level)

print(f"Confidence Interval ({confidence_level * 100}%): ({lower:.2f}, {upper:.2f})")
```

17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.


```python
import numpy as np
from scipy.stats import norm

def generate_data(mean, std_dev, size):
    """
    Generate random data from a normal distribution.

    Parameters:
        mean (float): Mean of the distribution.
        std_dev (float): Standard deviation of the distribution.
        size (int): Number of data points.

    Returns:
        numpy.ndarray: Generated data.
    """
    return np.random.normal(mean, std_dev, size)

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for the mean of the dataset.

    Parameters:
        data (array-like): Dataset.
        confidence_level (float): Desired confidence level (default is 95%).

    Returns:
        tuple: Lower and upper bounds of the confidence interval.
    """
    sample_mean = np.mean(data)
    std_error = np.std(data, ddof=1) / np.sqrt(len(data))
    z_value = norm.ppf((1 + confidence_level) / 2)
    margin_of_error = z_value * std_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return sample_mean, lower_bound, upper_bound

# Generate data
mean = 50  # Mean of the distribution
std_dev = 10  # Standard deviation of the distribution
size = 100  # Number of data points
data = generate_data(mean, std_dev, size)

# Calculate confidence interval
confidence_level = 0.95  # 95% confidence level
sample_mean, lower_bound, upper_bound = calculate_confidence_interval(data, confidence_level)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval ({confidence_level*100}%): ({lower_bound:.2f}, {upper_bound:.2f})")
```

18. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mean, std_dev):
    """
    Calculate and plot the PDF of a normal distribution.

    Parameters:
        mean (float): Mean of the normal distribution.
        std_dev (float): Standard deviation of the normal distribution.
    """
    # Generate values for the x-axis
    x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)  # Covering a wide range around the mean

    # Calculate the PDF using the normal distribution formula
    pdf = norm.pdf(x, loc=mean, scale=std_dev)

    # Plot the PDF
    plt.plot(x, pdf, color='blue', label=f'Normal PDF (mean={mean}, std={std_dev})')
    plt.fill_between(x, pdf, color='skyblue', alpha=0.3)  # Shading under the curve
    plt.title('Probability Density Function (PDF) of Normal Distribution')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.grid(alpha=0.7, linestyle='--')
    plt.legend()
    plt.show()

# Example usage
mean = 0  # Mean of the normal distribution
std_dev = 1  # Standard deviation of the normal distribution
plot_normal_pdf(mean, std_dev)
```

19. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.


```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

def plot_poisson_cdf(lambda_value, max_value):
    """
    Calculate and plot the CDF of a Poisson distribution.

    Parameters:
        lambda_value (float): The mean () of the Poisson distribution.
        max_value (int): The maximum value for plotting the CDF.
    """
    # Generate x values (discrete values for the Poisson distribution)
    x = np.arange(0, max_value + 1)

    # Calculate the CDF
    cdf = poisson.cdf(x, mu=lambda_value)

    # Plot the CDF
    plt.step(x, cdf, where='post', color='blue', label=f'Poisson CDF (={lambda_value})')
    plt.xlabel('Number of Events')
    plt.ylabel('Cumulative Probability')
    plt.title('Cumulative Distribution Function (CDF) of Poisson Distribution')
    plt.grid(alpha=0.7, linestyle='--')
    plt.legend()
    plt.show()

# Example usage
lambda_value = 5  # Mean () of the Poisson distribution
max_value = 20    # Maximum value for plotting
plot_poisson_cdf(lambda_value, max_value)
```

22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.


```python
import numpy as np

def calculate_range_and_iqr(data):
    """
    Calculate the range and interquartile range (IQR) of a dataset.

    Parameters:
        data (array-like): The dataset for which the range and IQR are calculated.

    Returns:
        tuple: Range (max - min) and IQR (Q3 - Q1).
    """
    data_range = np.max(data) - np.min(data)
    q1 = np.percentile(data, 25)  # First quartile (Q1)
    q3 = np.percentile(data, 75)  # Third quartile (Q3)
    iqr = q3 - q1  # Interquartile range

    return data_range, iqr

# Generate data from a normal distribution
mean = 50  # Mean of the normal distribution
std_dev = 10  # Standard deviation
size = 1000  # Number of data points
data = np.random.normal(mean, std_dev, size)

# Calculate the range and IQR
data_range, iqr = calculate_range_and_iqr(data)

print(f"Range of the dataset: {data_range:.2f}")
print(f"Interquartile Range (IQR) of the dataset: {iqr:.2f}")
```

23. Implement Z-score normalization on a dataset and visualize its transformation.


```python
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    """
    Apply Z-score normalization to a dataset.

    Parameters:
        data (array-like): Original dataset.

    Returns:
        numpy.ndarray: Z-score normalized data.
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    normalized_data = (data - mean) / std_dev
    return normalized_data

def visualize_transformation(original_data, normalized_data):
    """
    Visualize the transformation of a dataset after Z-score normalization.

    Parameters:
        original_data (array-like): Original dataset.
        normalized_data (array-like): Z-score normalized dataset.
    """
    plt.figure(figsize=(12, 6))

    # Original data histogram
    plt.subplot(1, 2, 1)
    plt.hist(original_data, bins=20, color='skyblue', edgecolor='blue', alpha=0.7)
    plt.title('Original Dataset')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(alpha=0.7, linestyle='--')

    # Normalized data histogram
    plt.subplot(1, 2, 2)
    plt.hist(normalized_data, bins=20, color='green', edgecolor='darkgreen', alpha=0.7)
    plt.title('Z-score Normalized Dataset')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.grid(alpha=0.7, linestyle='--')

    plt.tight_layout()
    plt.show()

# Example dataset
data = np.random.normal(loc=50, scale=10, size=1000)  # Generate a normal distribution

# Normalize the data using Z-score normalization
normalized_data = z_score_normalization(data)

# Visualize the transformation
visualize_transformation(data, normalized_data)
```

24.  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    """
    Calculate the skewness and kurtosis of a dataset.

    Parameters:
        data (array-like): The dataset for which skewness and kurtosis are calculated.

    Returns:
        tuple: Skewness and kurtosis of the dataset.
    """
    skewness = skew(data)  # Measure asymmetry of the distribution
    kurt = kurtosis(data)  # Measure the 'tailedness' of the distribution

    return skewness, kurt

# Generate data from a normal distribution
mean = 50  # Mean of the normal distribution
std_dev = 10  # Standard deviation
size = 1000  # Number of data points
data = np.random.normal(mean, std_dev, size)

# Calculate skewness and kurtosis
skewness, kurt = calculate_skewness_kurtosis(data)

print(f"Skewness of the dataset: {skewness:.4f}")
print(f"Kurtosis of the dataset: {kurt:.4f}")
```
